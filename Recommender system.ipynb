```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
df.info()
```

```python
movie_title = pd.read_csv("Movie_Id_Titles")
movie_title.head()
```

```python
df = pd.merge(df, movie_title, on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().nunique()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending = False).head(10)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending = False).head(10)
```

```python
ratings =  pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings
```

```python
ratings.head()
```

```python
ratings['count'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
plt.figure(figsize=(13,9))
ratings['count'].hist(bins=70)
```

```python
plt.figure(figsize=(13,9))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y ='count', data=ratings, alpha=0.4, height=8)
```

## Recomendando filmes semelhantes

```python
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat.shape
```

```python
ratings.sort_values('count', ascending=False).head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
```

```python
starwars_user_ratings
```

```python
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
liarliar_user_ratings
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars.head(10)
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
similar_to_liarliar.head(10)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['count'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['count']>100].sort_values('Correlation', ascending=False).head(100)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['count'])
```

```python
corr_liarliar[corr_liarliar['count']>100].sort_values('Correlation', ascending=False).head(100)
```
